```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
total_stats = pd.read_csv('data/totals_stats.csv')
print(total_stats.head())
```

```python
kobe = total_stats.loc[total_stats['Player'] == 'Kobe Bryant']
kobe_season = kobe.loc[kobe['RSorPO'] == 'Regular Season']
kobe_playoffs = kobe.loc[kobe['RSorPO'] == 'Playoffs']
```

```python
kobe['PPG'] = kobe['PTS']/kobe['G']
kobe['APG'] = kobe['AST']/kobe['G']
kobe['RPG'] = kobe['TRB']/kobe['G']
```

```python
fig, axes = plt.subplots(3, 1, figsize = (20, 15))


sns.lineplot(ax=axes[0], x='Season', y='PPG', data=kobe, marker='o', hue='RSorPO')
axes[0].set_title('Points per Game')
axes[0].set_xlabel('')
sns.lineplot(ax=axes[1], x='Season', y='APG', data=kobe, marker='o', hue='RSorPO')
axes[1].set_title('Assists per Game')
axes[1].set_xlabel('')
sns.lineplot(ax=axes[2], x='Season', y='RPG', data=kobe, marker='o', hue='RSorPO')
axes[2].set_title('Rebounds per Game')
axes[2].set_xlabel('')
plt.xlabel('Season')
```
